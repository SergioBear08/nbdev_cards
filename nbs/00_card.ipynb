```python
#| export
suits = ["♣️","♦️","♥️","♠️"]
ranks = [None, "A"] + [str(x) for x in range(2, 11)] + ["J", "Q", "K"]
```

We will be using numbers to represent playing cards clubs and ranks. These are the ranks:

```python
suits
```

For instance the suit at index `0` is:

```python
suits[1]
```

These are the ranks:

```python
ranks
```

For instance the rank at index `1`(note that there isn't a playing card at position `0`, since we want the ranks to match the indices where possible):

```python
ranks[1]
```

```python
#| export
class Card:
    "A playing card"
    def __init__(self,
                 suit,   # An index into `suits`
                 rank):  # An index into `ranks`
        self.suit, self.rank = suit, rank
    def __str__(self): return f"{suits[self.suit]}{ranks[self.rank]}"
    __repr__ = __str__
    def __eq__(self, b): return (self.suit, self.rank)==(b.suit, b.rank)
    def __lt__(self, b): return (self.suit, self.rank)<(b.suit, b.rank)
```

Here's an example of creating and displaying a card:

```python
c = Card(1, 3)
```

```python
print(c)
```

## Comparison operators

Equality, less than, and greater than work on the rank and suit indices:

```python
#| hide
test_eq(Card(1, 3), Card(1, 3))
test_ne(Card(1, 3), Card(2, 3))
test_ne(Card(1, 3), Card(1, 4))
```

```python
assert Card(1, 3) == Card(1, 3)
assert Card(1, 3) < Card(1, 4)
assert Card(2, 3) > Card(1, 3)
```
